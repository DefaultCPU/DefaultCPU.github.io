```python
print("1 - Add")
print("2 - Subtract")
print("3 - Multiply")
print("4 - Divide")
option = int(input("Choose an operation:"))

if(option in [1,2,3,4]):
    num1 = float(input("Enter first number:"))
    num2 = float(input("Enter second number:"))

    if(option == 1):
      result = num1 + num2
    elif(option == 2):
      result = num1 - num2
    elif(option == 3):
      result = num1 * num2
    elif(option == 4):
      result = num1 / num2 #Using // for division gives only whole integers while / with float on the num1 and 2 lines gives integers with decimals
else:
  print("Invalid option entered")

print("Result of operation is {}".format(result))
#This calculator works each for each function and has to be ran separately per function
```
